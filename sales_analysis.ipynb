```python
import pandas as pd
import numpy as np
```

```python
# Load dataset
sales_df = pd.read_csv('pdv_lineas.csv')
# Display dataframe
sales_df.head()
```

```python
# Verify data types
sales_df.dtypes
```

```python
#Convert date to datetime and rename column create_date to date
sales_df["create_date"] = sales_df["create_date"].astype('datetime64[ns]')
sales_df = sales_df.rename(columns={'create_date': 'date'})
sales_df.head()
```

```python
# Verify data types
sales_df.dtypes
```

```python
# Eliminating rows where create_uid cell, doesn't start with 'SUC'
new_sales_df= sales_df['create_uid'].str.startswith('SUC')
sales_df= sales_df.loc[new_sales_df,['id', 'order_id', 'date', 'product_id', 'create_uid',
                                                                'qty', 'price_unit', 'price_subtotal']]
sales_df.head()
```

```python
# Count if there is any 'False' values within our Df
false_df = (sales_df == 'False').sum().sum()
print(f"There are {false_df } 'False' values in the data frame.")
```

```python
new_df = sales_df['product_id'].str.split(', ', n=1, expand=True).rename(columns={0:'Product_ID', 1:'Description'})
new_df['Product_ID'] = new_df['Product_ID'].str.replace("[","")
new_df['Description'] = new_df['Description'].str.replace("[","")
new_df['Description'] = new_df['Description'].str.replace("]","")
new_df['Description'] = new_df['Description'].str.replace("'","")
new_df
```

```python
# Creating a new dataframe joining sales_df and new_df
new_sales_df = pd.concat([sales_df,new_df], axis = 1)
new_sales_df = new_sales_df.drop(columns = ['product_id', 'order_id'])
new_sales_df
```

```python
# Modifying branches names
new_sales_df['create_uid']= new_sales_df['create_uid'].replace({'SUC. ALLENDE (QRO)(2DO)': 'SUC. ALLENDE (QRO)',
                                                                'SUC. LEON CENTRO (LEON) (2DO)' : 'SUC. LEON CENTRO (LEON)',
                                                                'SUC. CORREGIDORA 4 (QRO) (2DO)' : 'SUC. CORREGIDORA 4 (QRO)',
                                                                'SUC. CORREGIDORA 2 (QRO) (2DO)' : 'SUC. CORREGIDORA 2 (QRO)'})
new_sales_df
```

```python
# Reorganize the order of our index - columns
new_sales_df = new_sales_df[['id', 'date', 'Product_ID', 'Description',
       'qty', 'create_uid', 'price_unit', 'price_subtotal']]
new_sales_df.head()
```

```python
# Define a list of words that represent devices
accesories_words = ['MICA', 'MICRO', 'SOPORTE', 'GLASS', 'SIM', 'TARJETA', 'TIEMPO AIRE', 'OTTER' , 
                    'ADAPTADOR DE CORRIENTE', 'USB', 'FLASH DRIVE', 'AUDIFONO', 'AUDÍFONOS', 'FDA',
                   'CARGADOR', 'PORTA', 'CABLE', 'TABLET', 'SD', 'ddd']

# Check which rows contain any of the accesories-related words
is_device = new_sales_df['Description'].str.contains('|'.join(accesories_words))

# Invert the boolean values to select only the rows where no accesories-related words are present
sales_devices_df = new_sales_df[~is_device]
# Displays dataframe with only cellphones
sales_devices_df
```

```python
# Export the final DataFrame to a CSV file
sales_devices_df.to_csv('ventas_contado_devices.csv', index=False)
```
