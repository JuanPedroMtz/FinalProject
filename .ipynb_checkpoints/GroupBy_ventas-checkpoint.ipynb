```python
# Import dependencies
import pandas as pd
import numpy as np
```

```python
ventas = pd.read_csv("./almacenes_ventas.csv")
ventas
ventas['month'] = pd.DatetimeIndex(ventas['date']).month
ventas_splited = pd.DataFrame(ventas.groupby(['date', 'month'])['qty'].sum().sort_values(ascending=False))
ventas_splited
```

```python
# Reaad CSV and convert to DataFrame 
ventas = pd.read_csv("./almacenes_ventas.csv")
ventas_brand = ventas['description'].str.split(" ", n = 1, expand = True)
ventas_brand = ventas_brand.rename(columns={0: 'brand', 1:'description'})
ventas = pd.concat([ventas, ventas_brand['brand']], axis=1)
ventas['month'] = pd.DatetimeIndex(ventas['date']).month
ventas = ventas.drop(columns = ['description', 'date'])
ventas = ventas[['product_id', 'name', 'price_unit', 'brand', 'month', 'qty']]
ventas
```

```python
# Groupby Product ID and Store
ventas_splited = pd.DataFrame(ventas.groupby(['name', 'month', 'price_unit', 'brand'])['qty'].sum().sort_values(ascending=False))
ventas_splited
```
