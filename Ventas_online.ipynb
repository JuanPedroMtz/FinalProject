```python
# Initial imports
import pandas as pd
import numpy as np
```

```python
# Load the ventas_linea.csv dataset.
ventas_linea_df = pd.read_csv('./ventas_linea.csv')
```

```python
ventas_linea_df. head()
```

```python
#Understand all the columns that exist in the data frame
print(ventas_linea_df.columns)
```

```python
# Keep all the branches that generate a transaction (SUC. (Sucursal)) on the create_uid column 
new_ventas_linea_df= ventas_linea_df['create_uid'].str.startswith('SUC')
ventas_linea_df= ventas_linea_df.loc[new_ventas_linea_df,['id',	'order_id',	'product_id',	'product_uom_qty',	'create_uid',	'price_unit',	'price_subtotal']]
ventas_linea_df.head()
```

```python
# Count if there is any 'False' values within our Df
false_ventas_linea_df = (ventas_linea_df == 'False').sum().sum()
print(f"There are {false_ventas_linea_df } 'False' values in the data frame.")
```

```python
# Drop all the 'False' values
ventas_linea_df = ventas_linea_df[ventas_linea_df != 'False'].dropna()
ventas_linea_df.head()
```

```python
# Double check that all 'False' values have been dropped.
false_ventas_linea_df = (ventas_linea_df == 'False').sum().sum()
print(f"There are {false_ventas_linea_df } 'False' values in the data frame.")
```

```python
# Split the 'product_id' column of ventas_linea_df into two columns using the comma delimiter and select each column
new_df = ventas_linea_df['product_id'].str.split(', ', n=1, expand=True).rename(columns={0:'Product_ID', 1:'Description'})
new_df['Product_ID'] = new_df['Product_ID'].str[1:]
new_df['Description'] = new_df['Description'].str[:-1]
new_df
```

```python
# Concatenate the original DataFrame with the new DataFrame
new_ventas_linea_df = pd.concat([ventas_linea_df.drop(columns=['product_id']), new_df], axis=1)

# Print the resulting DataFrame
new_ventas_linea_df
```

```python
# Reorganize the order of our index - columns
new_ventas_linea_df = new_ventas_linea_df.reindex(columns=['id', 'Product_ID', 'Description', 'order_id',
       'product_uom_qty', 'create_uid', 'price_unit', 'price_subtotal'])
new_ventas_linea_df.head()
```

```python
# Define a list of words that represent devices
accesories_words = ['MICA', 'MICRO', 'SOPORTE', 'GLASS', 'SIM', 'TARJETA', 'TIEMPO AIRE', 'OTTER' , 
                    'ADAPTADOR DE CORRIENTE', 'USB', 'FLASH DRIVE', 'AUDIFONO', 'AUDÍFONOS', 'FDA',
                   'CARGADOR', 'PORTA', 'CABLE','TABLET']

# Check which rows contain any of the accesories-related words
is_device = new_ventas_linea_df['Description'].str.contains('|'.join(accesories_words))

# Invert the boolean values to select only the rows where no accesories-related words are present
ventas_linea_devices_df = new_ventas_linea_df[~is_device]
```

```python
#Generate the dataframe showing only phones 
ventas_linea_devices_df 
```

```python
# Export the new DataFrame to a CSV file
ventas_linea_devices_df.to_csv('ventas_linea_devices.csv', index=False)
```
