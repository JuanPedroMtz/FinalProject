```python
# Import dependencies
import pandas as pd
import numpy as np4
```

```python
# Reaad CSV and convert to DataFrame 
products = pd.read_csv("./productos.csv")
products
```

```python
# Dropping N/A values 
products = products.dropna()
products['categ_id'].unique()

# No N/A values
```

```python
# Copying products dataframe
products1 = products.copy()
```

```python
# Filtering 'DISPOSITIVO / CELULAR / LIBRE' and 'DISPOSITIVO / CELULAR / TELCEL'
products = products[ (products['categ_id'] == 'DISPOSITIVO / CELULAR / LIBRE')]
products1 = products1[ (products1['categ_id'] == 'DISPOSITIVO / CELULAR / TELCEL')]
```

```python
# Concat dataframes and droping duplicates
frames = [products, products1]
products_clean = pd.concat(frames)
products_clean.drop_duplicates(['name'])
```

```python
# Verifying that filtering was done correctly
products_clean['categ_id'].unique()
```

```python
products_clean.rename(columns = {'product_id':'Product_ID'}, inplace = True)
products_clean.rename(columns = {'name':'Description'}, inplace = True)
products_clean
```

```python
products_clean = products_clean.reset_index(drop=True)
products_clean
```

```python
# Creating new CSV file
products_clean.to_csv('./productos_clean.csv', index=False)
```
